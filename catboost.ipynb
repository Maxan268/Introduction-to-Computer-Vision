```python
from catboost import CatBoostClassifier

best_model = CatBoostClassifier(
    iterations=995,
    learning_rate=0.06429082762539402,
    depth=6,
    l2_leaf_reg=0.9847976091102112,
    random_strength=0.0559357999518864,
    bagging_temperature=0.35409212726848965,
    border_count=136
)
```

```python
best_model.fit(X_train, y_train)
```

```python
# CatBooost
y_pred_cb= best_model.predict(X_test)
report_cb = classification_report(y_test.iloc[:, [0]], y_pred_cb)
```
